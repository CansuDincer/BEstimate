#### David Liu paper

- WT codon is *GAG* and the mutant one is *GTG*
- Mutation is on 5227002 and mutant sequence between 5227001-5227003 is CAC --> ABE --> CGC (+1 strand)

GCG --> Ala --> naturally occurring, non-sickling variant hemoglobin "Makassar" (HbG) [https://doi.org/10.1182/blood-2020-139016]

```python
uniprot_server = "https://www.ebi.ac.uk/proteins/api/"
uniprot_api = "proteins?offset=0&size=-1&accession=D9YZU5" 
api_request = requests.get(uniprot_server + uniprot_api, headers={"Accept": "application/json"})
hbb_seq = None
for i in api_request.json():
    hbb_seq = i["sequence"]["sequence"]

hbb_seq
```
